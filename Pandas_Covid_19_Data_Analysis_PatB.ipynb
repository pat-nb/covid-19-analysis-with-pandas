# Covid-19 Data Analysis
### This notebook is used to understand the comprehension of Data Analysis techniques using Pandas library.

### Data Source:
https://github.com/CSSEGISandData/COVID-19/tree/master/csse_covid_19_data/csse_covid_19_daily_reports

### File naming convention

MM-DD-YYYY.csv in UTC.

### Field description

- Province_State: China - province name; US/Canada/Australia/ - city name, state/province name; Others - name of the event (e.g., "Diamond Princess" cruise ship); other countries - blank.

- Country_Region: country/region name conforming to WHO (will be updated).

- Last_Update: MM/DD/YYYY HH:mm (24 hour format, in UTC).

- Confirmed: the number of confirmed cases. For Hubei Province: from Feb 13 (GMT +8), we report both clinically diagnosed and lab-confirmed cases. For lab-confirmed cases only (Before Feb 17), please refer to who_covid_19_situation_reports. For Italy, diagnosis standard might be changed since Feb 27 to "slow the growth of new case numbers." (Source)

- Deaths: the number of deaths.

- Recovered: the number of recovered cases.

#### Read the dataset

```python
import pandas as pd
covid_data = pd.read_csv("https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_daily_reports/11-01-2020.csv")
covid_data
```

#### Display the top 5 rows in the data

```python
covid_data.head(5)
```

#### Show the information of the dataset

```python
covid_data.info()
```

#### Show the sum of missing values of features in the dataset

```python
covid_data.isnull().sum()
```

#### Show the number of Confirmed cases by Country

```python
country_confirmed_sum = covid_data.groupby('Country_Region')['Confirmed'].sum().reset_index()
country_confirmed_sum
```

#### Show the number of Deaths by Country

```python
country_deaths_sum = covid_data.groupby('Country_Region')['Deaths'].sum().reset_index()
country_deaths_sum
```

#### Show the number of Recovered cases by Country

```python
country_recovered_sum = covid_data.groupby('Country_Region')['Recovered'].sum().reset_index()
country_recovered_sum
```

#### Show the number of Active Cases by Country

```python
country_active_sum = covid_data.groupby('Country_Region')['Active'].sum().reset_index()
country_active_sum
```

#### Show the latest number of Confirmed, Deaths, Recovered and Active cases Country-wise

```python
covid_data['Last_Update'] = pd.to_datetime(covid_data['Last_Update'])
latest_covid_data = covid_data.loc[covid_data.groupby('Country_Region')['Last_Update'].idxmax()]
latest_covid_data = latest_covid_data[['Country_Region', 'Confirmed', 'Deaths', 'Recovered', 'Active']]
latest_covid_data
```

### Show the countries with no recovered cases

```python
country_no_recovered = country_recovered_sum[country_recovered_sum['Recovered'] == 0]
country_no_recovered
```

#### Show the countries with no confirmed cases

```python
country_no_confirmed = country_confirmed_sum[country_confirmed_sum['Confirmed'] == 0]
country_no_confirmed
```

#### Show the countries with no deaths

```python
country_no_deaths = country_deaths_sum[country_deaths_sum['Deaths'] == 0]
country_no_deaths
```

```python
grouped_sum_data = covid_data.groupby('Country_Region').agg({
    'Confirmed': 'sum',
    'Deaths': 'sum',
    'Recovered': 'sum',
}).reset_index()

grouped_sum_data[grouped_sum_data['Deaths'] == 0]
```

#### Show the Top 10 countries with Confirmed cases

```python
grouped_sum_data = covid_data.groupby('Country_Region').agg({
    'Last_Update': 'max',
    'Confirmed': 'sum',
    'Deaths': 'sum',
    'Recovered': 'sum',
})

grouped_sum_data.sort_values(by='Confirmed', ascending=False).head(10)
```

#### Show the Top 10 Countries with Active cases

```python
grouped_sum_data = covid_data.groupby('Country_Region').agg({
    'Last_Update': 'max',
    'Confirmed': 'sum',
    'Deaths': 'sum',
    'Recovered': 'sum',
    'Active': 'sum'
})

grouped_sum_data.sort_values(by='Active', ascending=False).head(10)
```

#### Plot Country-wise Total deaths, confirmed, recovered and active casaes where total deaths have exceeded 50,000

```python
import matplotlib.pyplot as plt
```

```python
deaths_over_50000 = grouped_sum_data[grouped_sum_data['Deaths'] > 50000]

plt.plot(deaths_over_50000.index, deaths_over_50000['Confirmed'], label='Confirmed')
plt.plot(deaths_over_50000.index, deaths_over_50000['Recovered'], label='Recovered')
plt.plot(deaths_over_50000.index, deaths_over_50000['Active'], label='Active')

plt.title('Total Deaths(>50,000), Confirmed, Recovered and Active Cases by Country')
plt.legend()

plt.show()
```

### Plot Province/State wise Deaths in USA

```python
import plotly.express as px
```

```python
covid_data = pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_daily_reports/01-09-2021.csv')
```

```python
covid_data.columns
```

```python
us_data = covid_data[covid_data['Country_Region'] == 'US']

state_deaths_sum = us_data.groupby('Province_State')['Deaths'].sum().reset_index()

fig = px.bar(state_deaths_sum, x='Province_State', y='Deaths', title='State wise deaths reported of COVID-19 in USA', text_auto=True)
fig.update_xaxes(title='Provice_State')
fig.update_yaxes(title='Deaths')

fig.show()
```

### Plot Province/State Wise Active Cases in USA

```python
state_active_sum = us_data.groupby('Province_State')['Active'].sum().reset_index()

fig = px.bar(state_active_sum, x='Province_State', y='Active', title='State wise active reported of COVID-19 in USA', text_auto=True)
fig.update_xaxes(title='Provice_State')
fig.update_yaxes(title='Active')

fig.show()
```

### Plot Province/State Wise Confirmed cases in USA

```python
state_confirmed_sum = us_data.groupby('Province_State')['Confirmed'].sum().reset_index()

fig = px.bar(state_confirmed_sum, x='Province_State', y='Confirmed', title='State wise confirmed reported of COVID-19 in USA', text_auto=True)
fig.update_xaxes(title='Provice_State')
fig.update_yaxes(title='Confirmed')

fig.show()
```

```python
state_sum_data = covid_data.groupby('Province_State').agg({
    'Confirmed': 'sum',
    'Deaths': 'sum',
    'Recovered': 'sum',
    'Active': 'sum'
}).reset_index()

fig = fig = px.bar(state_sum_data, x='Province_State', y=['Confirmed', 'Deaths', 'Recovered', 'Active'],
             title='USA State wise combine number of confirmed, death, recovered, active COVID-19 cases',
             labels={'value': 'Count', 'variable': 'Case'})

fig.update_xaxes(title='Province_State')
fig.update_yaxes(title='Count')

fig.show()
```

### Plot Worldwide Confirmed Cases over time

```python
import plotly.express as px
import plotly.io as pio
```

```python
covid_data['Last_Update'] = pd.to_datetime(covid_data['Last_Update'])
covid_data['Last_Update_MY'] = covid_data['Last_Update'].dt.strftime('%b %Y')
worldwide_confirmed_sum = covid_data.groupby('Last_Update_MY')['Confirmed'].sum().reset_index()

fig = px.line(worldwide_confirmed_sum, x='Last_Update_MY', y='Confirmed',
              title='Worldwide Confirmed Novel Coronavirus(COVID-19) Cases Over Time',
              labels={'Last_Update_MY': 'Last_Update'})

pio.show(fig)
```
